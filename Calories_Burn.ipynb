```python
def rmse(target, predictions):
    return np.sqrt(np.mean(np.square(target-predictions)))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
inputs=df_cal[['Duration']]
outputs=df_cal.Calories
model.fit(inputs, outputs)
```

```python
predictions=model.predict(inputs)
predictions
```

```python
target=outputs
rmse(target, predictions)
```

```python
plt.scatter(inputs, target, alpha=0.5)
plt.plot(inputs, predictions, 'r-o', alpha=0.8)
plt.xlabel('Durations')
plt.ylabel('Calories_burn')
plt.show()
```

```python
inputs=df_cal[['Heart_Rate']]
output=df_cal.Calories
model.fit(inputs, outputs)
```

```python
predictions=model.predict(inputs)
print(np.min(predictions), np.max(predictions))
predictions
```

```python
target=output
rmse(target, predictions)
```

```python
plt.scatter(inputs, target, s=8, alpha=0.5)
plt.plot(inputs, predictions, 'r-o', alpha=0.8)
plt.xlabel('Heart_Rate')
plt.ylabel('Calories_burn')
plt.legend(['Actual', 'Prediction'])
plt.show()
```
